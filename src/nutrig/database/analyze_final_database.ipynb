```python
import os
import glob

import numpy as np

import matplotlib.pyplot as plt

import grand.dataio.root_trees as rt

import database.tools as tools

import importlib

#plt.style.use('seaborn-v0_8-colorblind')
plt.style.use('tableau-colorblind10')
plt.style.use('/pbs/home/p/pcorrea/tools/matplotlib_style_sans-serif.txt')

%matplotlib inline

tools.load_logger('debug')
```

```python
MODE             = 'UNIFORM'
BKG_DATABASE_DIR = '/sps/grand/pcorrea/nutrig/database/v2/bkg/lib'
BKG_FILE         = os.path.join(BKG_DATABASE_DIR,f'bkg_database_nutrig_v2_{MODE}.npz')
SIG_FILE         = BKG_FILE.replace('bkg','sig')

PLOT_TITLE = f'NUTRIG database v2, mode {MODE}'
```

```python
with np.load(BKG_FILE) as f:
    bkg_traces = f['traces']
    bkg_snr    = f['snr']

with np.load(SIG_FILE) as f:
    sig_traces = f['traces']
    sig_snr    = f['snr']
```

```python
snr_bins = np.linspace(3,8,6)
plt.hist(bkg_snr,label='Background',bins=snr_bins)
plt.hist(sig_snr,label='Signal',bins=snr_bins)
plt.xlabel('SNR')
plt.ylabel('Counts')
plt.title(PLOT_TITLE)
plt.legend(frameon=True)
plt.show()
```
